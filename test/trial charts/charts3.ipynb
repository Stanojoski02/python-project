```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
db1 = pd.read_excel("tbl_email_1.xlsx").head(10)
db2 = pd.read_excel("tbl_communication_stream_1.xlsx").head(50)
db3 = pd.read_excel("tbl_sentence_1.xlsx").head(50)
all_db = pd.concat([db1,db2], axis=1)
plt.figure(figsize=(50,40))
NG = nx.from_pandas_edgelist(db2, 'comm_stream_id', 'email_id')
values =[i * 1.05 for i in dict(NG.degree)]
color = []
for node in NG:
    if node in db2['comm_stream_id']:
        color.append("red")
    else:
        color.append("blue")
nx.draw(NG, with_labels = True, node_color=color, node_size=values)
```

```python
color = []
plt.figure(figsize=(20,10))
NG2 = nx.from_pandas_edgelist(db3, 'from(email index)', 'email_id')
for node in NG2:
    if node in db3['from(email index)']:
        color.append("red")
    else:
        color.append("blue")
nx.draw(NG2, with_labels = True, node_color=color)
```

```python
NG3 = nx.from_pandas_edgelist(db3, 'to(email index)', 'email_id')
plt.figure(figsize=(20,10))
color = []
for node in NG3:
    if node in db3['to(email index)']:
        color.append("red")
    else:
        color.append("blue")
nx.draw(NG3, with_labels = True, node_color=color)
```
