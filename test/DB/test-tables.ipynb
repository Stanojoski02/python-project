```python
import sqlite3
# Conect to the database or create it if it doesn't exist
conn = sqlite3.connect('example-01.db')
# Create a cursor object to execute SQL commands
c = conn.cursor()
# Create table with specified columns
c.execute('''CREATE TABLE tbl_email_address
             (ID INTEGER PRIMARY KEY, email TEXT)''')
# Insert data into the table
c.execute("INSERT INTO tbl_email_address VALUES (1, 'example1@email.com')")
c.execute("INSERT INTO tbl_email_address VALUES (2, 'example2@email.com')")
# Insert multiple records at once
data = [(3, 'example3@email.com'), (4, 'example4@email.com')]
c.executemany("INSERT INTO tbl_email_address VALUES (?,?)", data)
# Save changes to the database
conn.commit()
# Query data from the table
c.execute("SELECT * FROM tbl_email_address")
print(c.fetchall())
# Close the conection to the database
conn.close()
```
