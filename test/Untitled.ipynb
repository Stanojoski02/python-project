```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel('tbl_sentence_1.xlsx') # reads the excel file
sns.relplot(data=df[['verb','jaccard_similarity']],x="verb", y="jaccard_similarity", kind="line")
sns.relplot(data=df[['adjective','jaccard_similarity']],x="adjective", y="jaccard_similarity", kind="line")
```

```python
sns.relplot(data=df[['propn','jaccard_similarity']],x="propn", y="jaccard_similarity", kind="line")
sns.relplot(data=df[['noun','jaccard_similarity']],x="noun", y="jaccard_similarity", kind="line")
```
