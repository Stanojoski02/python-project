```python
import pandas as pd
import matplotlib.pyplot as plt

# увоз на ексел фајлот
df = pd.read_excel('tbl_sentence_0.xlsx').head(150)

# селекција на колоните што сакате да ги користите во чартот
x = df['ID']
y1 = df['cosine_similarity']
y2 = df['jaccard_similarity']
y3 = df["euclidean_distance"]
y4 = df["levenshtein_distance"]

# нарисување на чартот
plt.plot(x, y1, label='jaccard_similarity')
plt.plot(x, y2, label='cosine_similarity')
plt.xlabel('id')
plt.ylabel('Sentences')
plt.legend()
plt.show()

# нарисување на вториот чарт
plt.plot(x, y3, label='euclidean_distance')
plt.plot(x, y4, label='levenshtein_distance')
plt.xlabel('id')
plt.ylabel('distances')
plt.legend()
```

```python
x = df['ID']
y1 = df['adjective']
y2 = df['noun']

# нарисување на чартот
plt.bar(x, y1, label='adjective')
plt.bar(x, y2, label='noun')
plt.xlabel('id')
plt.ylabel('број на adjective и noun')
plt.legend()

plt.show()
```

```python
labels = ['propn', 'verb', 'adjective', 'noun']
sizes = [df['propn'].sum(), df['verb'].sum(), df['adjective'].sum(), df['noun'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')

plt.show()
```

```python
x = df['ID']
y1 = df['propn']
y2 = df['verb']

# нарисување на чартот
plt.scatter(x, y1, label='propn')
plt.scatter(x, y2, label='verb')
plt.xlabel('id')
plt.ylabel('Number of propn and verb')
plt.legend()

plt.show()
```

```python
labels = ['propn', 'verb', 'adjective', 'noun']
sizes = [df['propn'].sum(), df['verb'].sum(), df['adjective'].sum(), df['noun'].sum()]

# нарисување на чартот
plt.pie(sizes, labels=labels, autopct='%1.1f%%', radius=2.0)
plt.axis('equal')

plt.show()
```

```python
# селекција на филдовите
x = df['ID']
y1 = df['propn']
y2 = df['verb']

# нарисување на чартот
plt.plot(x, y1, label='propn')
plt.plot(x, y2, label='verb')
plt.xlabel('id')
plt.ylabel('број на propn и verb')
plt.legend()

plt.show()
```

```python
x = df['']
y1 = df['propn']
y2 = df['verb']
y3 = df['adjective']
y4 = df['noun']

# нарисување на чартот
plt.bar(x, y1 + y2 + y3 + y4, label='Sum of propn,verb, adjective, noun')
plt.xlabel('id')
plt.ylabel('број на propn, verb, adjective и noun')
plt.legend()

plt.show()
```
