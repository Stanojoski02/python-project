```python
import pandas as pd
import matplotlib.pyplot as plt

#    
df = pd.read_excel('tbl_sentence_0.xlsx').head(150)

#          
x = df['ID']
y1 = df['cosine_similarity']
y2 = df['jaccard_similarity']
y3 = df["euclidean_distance"]
y4 = df["levenshtein_distance"]

#   
plt.plot(x, y1, label='jaccard_similarity')
plt.plot(x, y2, label='cosine_similarity')
plt.xlabel('id')
plt.ylabel('Sentences')
plt.legend()
plt.show()

#    
plt.plot(x, y3, label='euclidean_distance')
plt.plot(x, y4, label='levenshtein_distance')
plt.xlabel('id')
plt.ylabel('distances')
plt.legend()
```

```python
x = df['ID']
y1 = df['adjective']
y2 = df['noun']

#   
plt.bar(x, y1, label='adjective')
plt.bar(x, y2, label='noun')
plt.xlabel('id')
plt.ylabel('  adjective  noun')
plt.legend()

plt.show()
```

```python
labels = ['propn', 'verb', 'adjective', 'noun']
sizes = [df['propn'].sum(), df['verb'].sum(), df['adjective'].sum(), df['noun'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')

plt.show()
```

```python
x = df['ID']
y1 = df['propn']
y2 = df['verb']

#   
plt.scatter(x, y1, label='propn')
plt.scatter(x, y2, label='verb')
plt.xlabel('id')
plt.ylabel('Number of propn and verb')
plt.legend()

plt.show()
```

```python
labels = ['propn', 'verb', 'adjective', 'noun']
sizes = [df['propn'].sum(), df['verb'].sum(), df['adjective'].sum(), df['noun'].sum()]

#   
plt.pie(sizes, labels=labels, autopct='%1.1f%%', radius=2.0)
plt.axis('equal')

plt.show()
```
